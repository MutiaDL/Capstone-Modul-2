### 1. Pendahuluan

SaaS (Software as a Service) merupakan suatu model bisnis di mana perangkat lunak disediakan sebagai layanan melalui internet, bukan dijual sebagai produk satuan. Dalam model ini, pengguna biasanya membayar biaya langganan secara berkala, baik bulanan maupun tahunan, untuk mengakses layanan dan fitur perangkat lunak yang dibutuhkan. Beberapa contoh populer dari layanan SaaS adalah Dropbox, Zoom, Google Workspace, dan Salesforce.

Model bisnis SaaS menawarkan fleksibilitas dan efisiensi yang tinggi, baik bagi penyedia layanan maupun pengguna, sehingga semakin banyak perusahaan yang mengadopsinya — terutama dalam skala B2B (Business to Business). #### (Cek Outlier Data Numerical)


```python
import pandas as pd

# Ambil kolom numerik
numerical_cols = df_cleaned.select_dtypes(include=['int64', 'float64']).columns

# Inisialisasi dataframe untuk menyimpan hasil
outlier_summary = pd.DataFrame(columns=["Kolom", "Jumlah Outlier", "Persentase Outlier (%)"])

# Cek outlier dengan metode IQR
for col in numerical_cols:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df_cleaned[(df_cleaned[col] < lower_bound) | (df_cleaned[col] > upper_bound)]
    num_outliers = outliers.shape[0]
    percent_outliers = round((num_outliers / df_cleaned.shape[0]) * 100, 2)

    outlier_summary = pd.concat([
        outlier_summary,
        pd.DataFrame([[col, num_outliers, percent_outliers]], columns=outlier_summary.columns)
    ], ignore_index=True)

# Tampilkan hasil
print(outlier_summary)
```

Dari hasil cek outlier diatas didapat outlier untuk data 'Sales' dan 'Profit' cukup besar diatas 10%. Tapi pada tahap ini data outlier tersebut tidak akan dihapus dulu karena bisa jadi dapat dijadikan analisis pada tahap analisis nanti.

Bisa jadi ini berasal dari kebijakan khusus, bug dalam sistem pricing, atau kondisi diskon kustom. Perlu perhatian apakah diskon-diskon ini efektif atau malah tidak terkontrol.

Selanjutnya hitung total sales dan total profit per nilai diskon → Apakah diskon yang besar justru membuat profit menurun?

```python
# Grouping berdasarkan diskon dan menghitung total sales & profit
diskon_agg = df_cleaned.groupby('Discount').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Order ID': 'count'  # jumlah transaksi
}).rename(columns={'Order ID': 'Transaction Count'})

# Hitung rata-rata profit per transaksi dan margin profit
diskon_agg['Avg Profit per Transaction'] = diskon_agg['Profit'] / diskon_agg['Transaction Count']
diskon_agg['Profit Margin (%)'] = (diskon_agg['Profit'] / diskon_agg['Sales']) * 100

# Sort berdasarkan diskon
diskon_agg = diskon_agg.sort_index()
display(diskon_agg)
```

Notes:

- Diskon 0% (Tanpa Diskon): Penjualan dan profit tertinggi ada di level diskon 0%, dengan sales sebesar 1.09 juta dan margin 29.5%. Jumlah transaksinya juga paling besar sebesar 4798yang artinya produk tanpa diskon justru berkontribusi paling besar ke profit. Hal ini menunjukkan bahwa tanpa diskon, bisnis masih mendapatkan keuntungan yang besar, yang mungkin menunjukkan bahwa produk sudah cukup kompetitif dengan harga asli.
- Diskon 10% hingga 20%: Di sini, meskipun profit dan total penjualan mengalami penurunan, average profit per transaction meningkat signifikan (terutama di diskon 10%). Ini menunjukkan bahwa ada sedikit penurunan dalam total keuntungan, tetapi keuntungan per transaksi mungkin lebih stabil atau lebih tinggi.
- Diskon 30% hingga 45%: Profit mulai menurun drastis, dan terdapat juga kerugian (- profit). Penurunan profit margin yang tajam pada diskon 30% hingga 45% menunjukkan bahwa diskon besar menyebabkan penurunan margin yang cukup signifikan dan produk mungkin sudah tidak menguntungkan lagi.
- Diskon 50% hingga 80%: Di level diskon ini, ada kerugian yang lebih besar pada penjualan, bahkan hingga -180%. Profit margin sangat negatif, yang berarti bisnis kehilangan banyak uang dalam setiap transaksi. Dengan diskon sebesar ini, keuntungan lebih kecil dari biaya atau harga produk, yang jelas menunjukkan bahwa strategi diskon besar-besaran ini tidak menguntungkan.

Untuk Menyederhanakan variasi diskon yang terlalu banyak (kayak 0.32, 0.45, 0.7, dst.) dan agar lebih mudah untuk membandingkan kinerja antar grup maka dilakukan discount grouping atau segmentasi diskon. Pada tahap ini segmentasi diskon akan dibagi menjadi 4 kelompok yaitu: kelompok tanpa diskon (diskon 0%), kelompok diskon ringan (diskon 10-20%), kelompok diskon sedang (diskon 30-50%), dan kelompok diskon tinggi (>50%).

```python
# Menambahkan kolom kategori diskon
def categorize_discount(x):
    if x == 0:
        return 'Tanpa Diskon'
    elif 0 < x <= 0.20:
        return 'Diskon Ringan (10-20%)'
    elif 0.20 < x <= 0.50:
        return 'Diskon Sedang (30-50%)'
    else:
        return 'Diskon Tinggi (>50%)'

df_cleaned['Diskon_Kategori'] = df_cleaned['Discount'].apply(categorize_discount)

# Grup dan agregasi data
grouped_discount = df_cleaned.groupby('Diskon_Kategori').agg(
    Total_Sales=('Sales', 'sum'),
    Total_Profit=('Profit', 'sum'),
    Jumlah_Transaksi=('Order ID', 'count')
)

# Hitung rata-rata per transaksi dan margin
grouped_discount['Rata-rata Sales/Transaksi'] = grouped_discount['Total_Sales'] / grouped_discount['Jumlah_Transaksi']
grouped_discount['Rata-rata Profit/Transaksi'] = grouped_discount['Total_Profit'] / grouped_discount['Jumlah_Transaksi']
grouped_discount['Profit Margin (%)'] = (grouped_discount['Total_Profit'] / grouped_discount['Total_Sales']) * 100

# Urutkan berdasarkan kategori diskon
grouped_discount = grouped_discount.sort_index()

display(grouped_discount)
```

1. Tanpa Diskon

Ini merupakan kategori dengan penjualan tertinggi dan margin tertinggi. Tanpa diskon, perusahaan tetap mampu menghasilkan penjualan besar dan profit yang sangat baik. Ini indikasi bahwa produk cukup kuat tanpa perlu banyak insentif diskon.

2. Diskon Ringan (10–20%)

Memberikan diskon ringan masih cukup efektif dalam mendorong penjualan tanpa mengorbankan profit terlalu besar. Ini bisa dianggap sebagai strategi kompromi.

3. Diskon Sedang (30–50%)

Walaupun rata-rata sales per transaksi tinggi (556), tapi margin sangat buruk dan profit justru rugi.

4. Diskon Tinggi (>50%)

Ini menunjukkan strategi diskon tinggi justru membuat perusahaan rugi besar.Rata-rata profit per transaksi juga sangat negatif, padahal transaksi cukup banyak (856).

Kesimpulan Sementara
- Strategi tanpa diskon atau diskon ringan (maksimal 20%) adalah yang paling sehat dari sisi profit.
- Diskon besar (>30%) berisiko menyebabkan kerugian, bahkan saat volume penjualan meningkat.

#### **Profit Margin per Produk**

Untuk menilai efisiensi keuntungan tiap produk terhadap penjualannya, sehingga kita bisa tahu produk mana saja yang mempunyai margin tinggi yang berpotensi memeberikan keuntungan besar bahkan saat volume rendah.

```python
profit_margin_df = df_cleaned.groupby('Product').agg({
    'Profit': 'sum',
    'Sales': 'sum'
}).reset_index()

profit_margin_df['Profit_Margin'] = profit_margin_df['Profit'] / profit_margin_df['Sales']
profit_margin_df = profit_margin_df.sort_values(by='Profit_Margin', ascending=False)
profit_margin_df.reset_index(drop=True, inplace=True)

display(profit_margin_df)
```

* Profit Margin menunjukkan berapa besar keuntungan yang diperoleh dari setiap penjualan produk. Contoh dari data diatas:
    - Alchemy: Margin = 0.37 → Artinya dari setiap 1 unit uang penjualan, didapatkan 0.37 unit keuntungan (37% margin).
    - SaaS Connector Pack - Gold: Margin = 0.42 → Artinya ini produk paling efisien dalam menghasilkan profit. Untuk setiap 1 unit penjualan, dia menghasilkan 0.42 profit (42%).
* Produk dengan margin tinggi bisa dianggap efisien dan berpotensi diprioritaskan, apalagi kalau juga punya volume penjualan tinggi. 
* Produk dengan margin rendah atau negatif bisa jadi boros biaya atau tidak layak dipertahankan tanpa intervensi strategi.

```python
# Visualisasi Profit Margin per Produk 
plt.figure(figsize=(10, 6))
barplot = sns.barplot(
    data=profit_margin_df, 
    y='Product', 
    x='Profit_Margin', 
    palette='coolwarm'
)

# label pada setiap batang
for i, row in profit_margin_df.iterrows():
    plt.text(row['Profit_Margin'] + 0.005, i, f"{row['Profit_Margin']:.2f}", va='center')

plt.title('Profit Margin per Produk')
plt.xlabel('Profit Margin')
plt.ylabel('Produk')
plt.tight_layout()
plt.show()
```

Dari hasil pengecekan efisiensi penjualan diatas, misal kita bagi menjadi 4 kelompok produk, maka:

#### **Segmentasi Produk Berdasarkan Profitabilitas**

Setelah dilakukan analisis sebelumnya (total profit per produk, profit margin per produk, rata-rata profit per transaksi), maka untuk mengevaluasi kontribusi profit produk secara komprehensif maka produk dikelompokkan ke dalam segmen berdasarkan beberapa metrik sekaligus dengan ketentuan sbb:
1. Bintang (Star) -> Total profit tinggi, margin tinggi, avg profit tinggi
2. Volume Driver -> Total profit tinggi, margin sedang, avg profit sedang
3. Premium Niche -> Total profit rendah, tapi margin dan avg profit tinggi
4. Loss Maker ->Total profit dan margin rendah, bahkan negatif

```python
# Gabungkan data dari tahap sebelumnya
segment_df = avg_profit_df.copy()
segment_df = segment_df.merge(profit_margin_df[['Product', 'Profit_Margin']], on='Product')

# Hitung nilai rata-rata untuk pembanding segmentasi
avg_total_profit = segment_df['Total_Profit'].mean()
avg_profit_margin = segment_df['Profit_Margin'].mean()
avg_avg_profit = segment_df['Avg_Profit_per_Transaction'].mean()

# Fungsi segmentasi
def classify_segment(row):
    if row['Total_Profit'] > avg_total_profit and row['Profit_Margin'] > avg_profit_margin and row['Avg_Profit_per_Transaction'] > avg_avg_profit:
        return 'Star'
    elif row['Total_Profit'] > avg_total_profit:
        return 'Volume Driver'
    elif row['Profit_Margin'] > avg_profit_margin and row['Avg_Profit_per_Transaction'] > avg_avg_profit:
        return 'Premium Niche'
    else:
        return 'Loss Maker'

# Terapkan fungsi ke data
segment_df['Segment'] = segment_df.apply(classify_segment, axis=1)

# Lihat hasil segmentasi
display(segment_df[['Product', 'Total_Profit', 'Profit_Margin', 'Avg_Profit_per_Transaction', 'Segment']].sort_values(by='Segment').reset_index(drop=True))
```
